```python
 # from the above boxplot diagram it is clear that there we found some outliers in the datadet, so that it is necessary that we need to deal with the outliers.
```

```python
for column in x.columns:
    q1=x[column].quantile(0.25)
    q3=x[column].quantile(0.75)
    iqr=q3-q1
    lower_bound=q1-1.5*iqr
    upper_bound=q3+1.5*iqr
    mask=(x[column]>=lower_bound)&(x[column]<=upper_bound)
```

```python
#Now mask the outliers present in the dataset.
x_outlier_detection=x[mask]
y_outlier_detection=y[mask]
```

```python
#Upto here we have performed the Exploratory Data Analysis part and now its the turn to train our Machine Learning model
```

```python
# Now we will split our data into train and test.
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(newdata, y_outlier_detection, test_size=0.2, random_state=42)
```

```python
y_train.value_counts()
```

```python
#As we can clearly see there is alot of data imbalance in the training data, as these can affec the model accuracy and the prediction. So we definetly need to address this case as it is a health care problem.
#Addressing the data imbalance is can be done in three different processes. They are as OverSampling, UnderSampling and SMOTE techniques.
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
x_train_resampled, y_train_resampled = smote.fit_resample(x_train,y_train)
```

```python
y_train_resampled.value_counts()
```

```python
#Now it is clear that we have the proper balanced data to train our machine learning model on implementing the algorithum.
```

```python
from sklearn.linear_model import LogisticRegression
classification = LogisticRegression()
classification.fit(x_train_resampled, y_train_resampled)
```

```python
y_predictions = classification.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_predictions, y_test)
```
